```python
import os
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
pd.set_option('display.max_columns', 1000)


with open('Core_set/LUSC/lusc_mirna_core.txt') as f:
    ct = 0
    for i in f.readlines():
        if ct  == 0:
            col_kirc = pd.Series(i.strip().split('\t'))
            df_kirc_mirna = pd.DataFrame(columns=col_kirc)
        else:
            df_tmp = pd.DataFrame([dict(zip(col_kirc, i.strip().split('\t')))])
            df_kirc_mirna = pd.concat([df_kirc_mirna, df_tmp], axis=0)
        ct+=1

with open('Core_set/LUSC/lusc_mrna_core.txt') as f:
    ct = 0
    for i in f.readlines():
        if ct  == 0:
            col_kirc = pd.Series(i.strip().split('\t'))
            df_kirc_mrna = pd.DataFrame(columns=col_kirc)
        else:
            df_tmp = pd.DataFrame([dict(zip(col_kirc, i.strip().split('\t')))])
            df_kirc_mrna = pd.concat([df_kirc_mrna, df_tmp], axis=0)
        ct+=1
        
with open('Core_set/LUSC/lusc_clinical_core.txt') as f:
    ct = 0
    for i in f.readlines():
        if ct  == 0:
            col_kirc = pd.Series(i.strip().split('\t'))
            df_kirc_clinical = pd.DataFrame(columns=col_kirc)
        else:
            df_tmp = pd.DataFrame([dict(zip(col_kirc, i.strip().split('\t')))])
            df_kirc_clinical = pd.concat([df_kirc_clinical, df_tmp], axis=0)
        ct+=1

with open('Core_set/LUSC/lusc_rppa_core.txt') as f:
    ct = 0
    for i in f.readlines():
        if ct  == 0:
            col_kirc = pd.Series(i.strip().split('\t'))
            df_kirc_rppa = pd.DataFrame(columns=col_kirc)
        else:
            df_tmp = pd.DataFrame([dict(zip(col_kirc, i.strip().split('\t')))])
            df_kirc_rppa = pd.concat([df_kirc_rppa, df_tmp], axis=0)
        ct+=1
        
with open('Core_set/LUSC/lusc_os_core.txt') as f:
    ct = 0
    for i in f.readlines():
        if ct  == 0:
            col_kirc = pd.Series(i.strip().split('\t'))
            df_kirc_os = pd.DataFrame(columns=col_kirc)
        else:
            df_tmp = pd.DataFrame([dict(zip(col_kirc, i.strip().split('\t')))])
            df_kirc_os = pd.concat([df_kirc_os, df_tmp], axis=0)
        ct+=1
        
with open('Core_set/LUSC/lusc_cnv_core.txt') as f:
    ct = 0
    for i in f.readlines():
        if ct  == 0:
            col_kirc = pd.Series(i.strip().split('\t'))
            df_kirc_cnv = pd.DataFrame(columns=col_kirc)
        else:
            df_tmp = pd.DataFrame([dict(zip(col_kirc, i.strip().split('\t')))])
            df_kirc_cnv = pd.concat([df_kirc_cnv, df_tmp], axis=0)
        ct+=1
        
df_kirc = pd.merge(df_kirc_mirna, df_kirc_mrna, on='feature', how='left')
df_kirc = pd.merge(df_kirc, df_kirc_clinical, on='feature', how='left')
df_kirc = pd.merge(df_kirc, df_kirc_rppa, on='feature', how='left')
df_kirc = pd.merge(df_kirc, df_kirc_os, on='feature', how='left')
df_kirc = pd.merge(df_kirc, df_kirc_cnv, on='feature', how='left')
df_kirc.to_csv('df_lusc.csv')
```

## Add label

```python
import pandas as pd

projects = ['ov', 'kirc', 'lusc', 'gbm']
for project in projects:
    df = pd.read_csv('df_%s.csv' %project, index_col='Unnamed: 0')

    with open('Binary/%s_binary_survival.txt' %project) as f:
        ct = 0
        for i in f.readlines():
            if ct  == 0:
                col = pd.Series(i.strip().split('\t'))
                df_ans = pd.DataFrame(columns=col)
            else:
                df_tmp = pd.DataFrame([dict(zip(col, i.strip().split('\t')))])
                df_ans = pd.concat([df_ans, df_tmp], axis=0)
            ct+=1
    df_ans.head()
    print(len(df_ans))
    
    df = pd.merge(df, df_ans, on='feature', how='left')
    df.to_csv('df_%s_with_ans.csv' %project)
```

## Split into Train and Test

```python
import pandas as pd

df_ov = pd.read_csv('df_ov_with_ans.csv', index_col=['Unnamed: 0'])
df_kirc = pd.read_csv('df_kirc_with_ans.csv', index_col=['Unnamed: 0'])
df_lusc = pd.read_csv('df_lusc_with_ans.csv', index_col=['Unnamed: 0'])
df_gbm = pd.read_csv('df_gbm_with_ans.csv', index_col=['Unnamed: 0'])
```

```python
project = 'ov'

with open('%s_train_sample_list.txt' %project) as f:
    list_train = f.read().split()
    print(len(list_train))
    
with open('%s_test_sample_list.txt' %project) as f:
    list_test = f.read().split()
    print(len(list_test))
    
df_train = pd.DataFrame({'feature':pd.Series(list_train)})
df_test = pd.DataFrame({'feature':pd.Series(list_test)})

df_train_tmp = pd.merge(df_train, df_ov, on='feature', how='inner')
df_train_tmp = df_train_tmp.groupby(by='feature', as_index=False).first()

df_train_tmp[df_train_tmp['is_alive'].isna() == False]
df_train_tmp[df_train_tmp['is_alive'].isna() == True]

df_test_tmp = pd.merge(df_test, df_ov, on='feature', how='inner')
df_test_tmp = df_test_tmp.groupby(by='feature', as_index=False).first()

print(len(df_test_tmp[df_train_tmp['is_alive'].isna() == False]))
print(len(df_test_tmp[df_train_tmp['is_alive'].isna() == True]))
```
